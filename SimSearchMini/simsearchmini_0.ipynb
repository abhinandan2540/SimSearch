```python
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu' # setting up device for model training
print(device)
```

```python
# importing stuffs for model training
from models.model import SimSearchMiniModel # this inlcude both encoders and projections
from loss.nt_xnt import nt_xnt_loss
from data.dataset import SimSearchMiniDataset
from utils.transformation import SimSearchMini_transform
```

```python
# creating dataloader for feeding data into model 
from torch.utils.data import DataLoader
dataset=SimSearchMiniDataset(root_dir="dataset", transform=SimSearchMini_transform) # opening dataset
loader=DataLoader(dataset=dataset, batch_size=32, shuffle=True) # creating dataloader
```

```python
# opening model
simsearchminimodel=SimSearchMiniModel()

# adamw optimizer function
optimizer = torch.optim.AdamW(
    simsearchminimodel.parameters(),
    lr=3e-4,
    weight_decay=1e-4
)
```

```python
# SimSearchmini model training
def SimSearchMini_training(model, loader, loss_fn, optimizer_fn, device, epochs):

    for epoch in range(epochs):
        for view1, view2 in loader:  # iterating through loader
            # loading view1 into device, similarly for view2
            view1 = view1.to(device)
            view2 = view2.to(device)

            z1 = model(view1)  # embedding for view1
            z2 = model(view2)  # embedding for view2

            loss = loss_fn(z1, z2)  # calculating loss
            optimizer_fn.zero_grad()  # zero gradient of optimizer
            loss.backward()  # backpropagation
            optimizer_fn.step()  # usign optimizer function
        print(f"Epoch: {epoch}| Loss: {loss.item()}")


SimSearchMini_training(model=simsearchminimodel, loader=loader, loss_fn=nt_xnt_loss,
                   optimizer_fn=optimizer, device=device, epochs=50)

# training needs a good compute power
# need something for it
```
