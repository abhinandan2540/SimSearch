data augmentation to balance unbalanced dataset

```python
from PIL import Image
from pathlib import Path

# images inside dataset can be of many types
# png, jpg, jpeg, gif, transparent
# for training self-supervised model rgb image is good
# cleaning all types of images into rgb image for good self supervised model training
def clean_dataset(dataset_path):
    dataset_path = Path(dataset_path)
    for img_path in dataset_path.rglob("*"):
        if img_path.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
            continue
        try:
            img = Image.open(img_path)
            if img.mode != "RGB":
                img = img.convert("RGB")
                img.save(img_path)

        except Exception as e:
            print("Skipping:", img_path, e)


clean_dataset("mammals")
```

```python
import Augmentor
import os
from tqdm import tqdm 
from pathlib import Path

def data_augmentation(data_path:str, target_img_count:int):
    data_path=Path(data_path)
    for folder in os.listdir(data_path):

        class_path=data_path/folder # each class path inside the mamals data_path
        
        if not class_path.is_dir(): # if class_path is not a directory then continue
            continue

        file_images=[f for f in os.listdir(class_path) if os.path.isfile(class_path/f) and not f.startswith('.')]
        current_img_count=len(file_images)
        generate_img_count=target_img_count-current_img_count

        if current_img_count>=target_img_count:  # folder has already enough images as specified in target
            return
        if generate_img_count<=0: # also the same condition
            return
        P=Augmentor.Pipeline(source_directory=str(class_path), output_directory='augmented_output')
        P.zoom_random(probability=0.5, percentage_area=0.9)
        P.flip_left_right(probability=0.5)
        

        print(f"\nclass: {folder}")
        print(f"original count: {current_img_count}")
        print(f"generating: {generate_img_count}")
        print(f"target count: {target_img_count}")

        P.sample(generate_img_count)

        print(f"augmentated {folder}")




data_path="mammals"
target_img_count=500
data_augmentation(data_path=data_path, target_img_count=target_img_count)

```
