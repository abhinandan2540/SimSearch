```python
# DON'T RUN THIS TWICE

from PIL import Image
from pathlib import Path

# images inside dataset can be of many types
# png, jpg, jpeg, gif, transparent
# for training self-supervised model rgb image is good
# cleaning all types of images into rgb image for good self supervised model training
def clean_dataset(dataset_path):
    dataset_path = Path(dataset_path)
    for img_path in dataset_path.rglob("*"):
        if img_path.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
            continue
        try:
            img = Image.open(img_path)
            if img.mode != "RGB":
                img = img.convert("RGB")
                img.save(img_path)

        except Exception as e:
            print("Skipping:", img_path, e)


clean_dataset("data")
```
