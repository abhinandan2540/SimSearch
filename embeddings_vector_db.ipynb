```python
# loading the saved embeddings
import torch
data=torch.load("SimSearch_0_image_embeddings.pt")
embeddings=data["SimSearch_0_embeddings"]
paths=data["SimSearch_0_paths"] # paths are the image paths
```

```python
# for vector db
# faiss: facebook ai similarity search
import faiss
import numpy as np

embedding_dim=128
index=faiss.IndexFlatL2(embedding_dim)

# print(np.array(embeddings).shape) # initial shape of embeddings (4377,1,128)
# faiss works on 2 params (images, dimentional embeddings)
embeddings_np = np.array(embeddings).squeeze().astype("float32") # (4377,128)


index = faiss.IndexFlatL2(embedding_dim) # index of the embedding dimentions
index.add(embeddings_np)
print(index.ntotal) # total number of vectors in index, it represents the number of images of which embeddings has been stored

```

```python
import pickle

# saving the faiss index 
faiss.write_index(index, "SimSearch_0_image_index.faiss")
# saving image paths
with open("SimSearch_0_image_paths.pkl","wb") as f:
    pickle.dump(paths, f)
```
